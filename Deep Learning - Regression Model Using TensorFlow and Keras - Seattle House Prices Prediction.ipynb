## 3. Separating features from labels

```python
X = df.drop('price', axis=1).values  # .values - returning back NumPy array, to make sure there are no errors between numpy and pandas datatypes.
y = df['price'].values
```

## 4. Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Scaling the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## 5. Creating the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
```

```python
X_train.shape
```

```python
# Since we have 19 features, we add 19 neurons in a layer
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))  # Output layer

model.compile(optimizer='adam', loss='mse')
```

```python
# Passing in X and y, also validation data. Means after each epoch, the code will run the test data and check the loss. To keep a tracking.
# batch_size should be in power of 2, 64, 128, 256. Training the data in smaller batches makes the model better.

# Training the model
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), batch_size=128, epochs=400)
```

## 6. Model evaluation

```python
losses = pd.DataFrame(model.history.history)
# These are the losses
```

```python
losses
# val_loss is the loss on the test set.
```

```python
losses.plot()
# Good to have the 2 lines together. This is too much overfitting b/c the 2 lines are not together
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
# We're off $100k on avg
```

```python
df['price'].describe()

# We're off by 20%
```

```python
explained_variance_score(y_test, predictions)

# best score is 1.0
```
