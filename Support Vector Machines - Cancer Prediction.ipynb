# Exploratory data analysis 

```python
# Write a code that could help me visualize this df_feat dataset 
# Python
# import necessary libraries


# Calculate correlations
corr = df_feat.corr()

# Using a heatmap to visualize the correlation matrix
plt.figure(figsize=(14,14))  #14in by 14in
sns.heatmap(corr, cbar = True, square = True, annot=True, fmt= '.1f', 
            annot_kws={'size': 15}, cmap='Greens')
plt.title('Correlation between features')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Get our train test split
X = df_feat
y = cancer['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=101)

model = SVC()
model.fit(X_train, y_train)
model_preds = model.predict(X_test)
confusion_matrix(y_test, model_preds)
print(classification_report(y_test, model_preds))
print('\n')
print(confusion_matrix(y_test, model_preds))
```

# Perform grid search
Find the best params setting, by changing the 'gamma' and 'C' values

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100,1000], 'gamma':[1,0.1,0.01,0.001,0.0001]}
grid = GridSearchCV(SVC(), param_grid, verbose=3)
grid.fit(X_train, y_train)
```

```python
grid.best_params_   # Best params
```

```python
# Rerun the model
grid_predictions = grid.predict(X_test)
print(confusion_matrix(y_test, grid_predictions))
print('\n')
print(classification_report(y_test, grid_predictions))
```

```python
# Classification report
best_state = 0
max_accuracy = 0
best_report = None

X = df_feat
y = cancer['target']

for random_state in range(1,101):
    model = SVC(C = 1, gamma = 0.0001)   # put in the new C and gamma values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_state)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    report = classification_report(y_test, predictions, output_dict=True)
    accuracy = accuracy_score(y_test, predictions)
    if accuracy > max_accuracy:
        max_accuracy = accuracy
        best_state = random_state
        best_report = report

print('Best random state: ', best_state)
print('Classification Report: ')
print(pd.DataFrame(best_report).transpose())
```
